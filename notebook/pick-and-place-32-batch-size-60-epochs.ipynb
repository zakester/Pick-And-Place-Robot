```python
import numpy as np
import os

from tflite_model_maker.config import ExportFormat, QuantizationConfig
from tflite_model_maker import model_spec
from tflite_model_maker import object_detector

from tflite_support import metadata

import tensorflow as tf
assert tf.__version__.startswith('2')

tf.get_logger().setLevel('ERROR')
from absl import logging
logging.set_verbosity(logging.ERROR)
```

```python
train_data = object_detector.DataLoader.from_pascal_voc(
    '../input/pick-and-place-packages-mobile-robot/dataset/dataset/train',
    '../input/pick-and-place-packages-mobile-robot/dataset/dataset/train',
    ['zone', 'package', 'package_target']
)

val_data = object_detector.DataLoader.from_pascal_voc(
    '../input/pick-and-place-packages-mobile-robot/dataset/dataset/validation',
    '../input/pick-and-place-packages-mobile-robot/dataset/dataset/validation',
    ['zone', 'package', 'package_target']
)
```

```python
spec = model_spec.get('efficientdet_lite2')
```

```python
model = object_detector.create(train_data, model_spec=spec, batch_size=32, train_whole_model=True, epochs=60, validation_data=val_data)
```

```python
model.evaluate(val_data)
```

```python
model.export(export_dir='.', tflite_filename='pick_place_32b_60e.tflite')
```

```python
model.evaluate_tflite('pick_place_32b_60e.tflite', val_data)
```
